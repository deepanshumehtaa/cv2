```python
#!pip install imutils
import numpy as np
import imutils
import cv2

%matplotlib inline

# Imutils are a series of convenience functions to make basic img processing 
# functions such as translation, rotation, resizing, skeletonization  and displaying Matplotlib images easier with OpenCV
```

```python
dir = 'car.jpg'
img = cv2.imread(dir)  # image type is nparray

cv2.imshow('gadi', img)  # gadi is the title
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
#import matplotlib.image as mpimg

plt.figure(figsize=(8,8))
plt.axis("off")
imgplot = plt.imshow(img)
```

we have seen both are the same images but of diffrent colour

becoz, OpenCV represents RGB images as multi-dimensional NumPy arrays…but in reverse order!

This means that images are actually represented in *BGR* order rather than *RGB!*

## **convert the image from BGR to RGB**


```python
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

# here we always use plt to show image and not cv2
```

## Denoising

```python
#denoising
img_ = cv2.GaussianBlur(img, (3,3), 0)
gau_img = cv2.fastNlMeansDenoisingColored(img_, None, 10, 10, 7, 15)  # 10,10 are wieghts
                                                                     # 7,15 intensity of filter, intensity of rgb filter

plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(cv2.cvtColor(gau_img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Edge detection

```python
# Canny

edges_canny = cv2.Canny(img,100,150)

plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(cv2.cvtColor(edges_canny, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Laplacian Image 

ddepth = cv2.CV_16S
lap_img = cv2.Laplacian(img, ddepth, ksize=3)


plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(lap_img)
plt.show()
```

## sharpening the edges

```python
#ddepth = cv2.CV_16S
ddepth = cv2.CV_8U

lap_img = cv2.Laplacian(img, ddepth, ksize=3)

sharp_img = img + lap_img  + lap_img


gau_img = cv2.fastNlMeansDenoisingColored(lap_img, None, 10, 10, 30, 15)


plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(gau_img)
plt.show()

#not so much diffrence
```
